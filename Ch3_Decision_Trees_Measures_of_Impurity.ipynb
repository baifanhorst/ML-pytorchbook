```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Visualizing the entropy for two classes
# It can be seen that the entropy is maximum when the proportion of each class is 1/2, i.e., uniform distribution
def entropy(p):
    q = 1-p
    if p==1 or q==1:
        return 0
    else: 
        return -p * np.log2(p) - q * np.log2(q)
    
x_list = np.linspace(0, 1, 1001)
ent = np.zeros(x_list.shape)
for i in range(x_list.shape[0]):
    ent[i] = entropy(x_list[i])

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x_list, ent)

ax.set_xlabel("Proportion of Class 1")
ax.set_ylabel("Entropy")
```

```python
# Comparison of different impurity measures
def gini(p):
    q = 1-p
    return 1 - p**2 - q**2

def classification_error(p):
    q = 1-p
    return 1 - np.max([p,q])

x_list = np.linspace(0, 1, 1001)
ent = np.zeros(x_list.shape)
gi = np.zeros(x_list.shape)
ce = np.zeros(x_list.shape)

for i in range(x_list.shape[0]):
    ent[i] = entropy(x_list[i])
    gi[i] = gini(x_list[i])
    ce[i] = classification_error(x_list[i])

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x_list, ent, label="Entropy")
ax.plot(x_list, ent/2, label="Entropy rescaled")
ax.plot(x_list, gi, label="Gini")
ax.plot(x_list, ce, label="Classification error")

ax.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=5, fancybox=True, shadow=False)

ax.axhline(y=0.5, linewidth=1, color='k', linestyle="--")
ax.axhline(y=1, linewidth=1, color='k', linestyle="--")


ax.set_xlabel("Proportion of Class 1")
ax.set_ylabel("Measures")
```
